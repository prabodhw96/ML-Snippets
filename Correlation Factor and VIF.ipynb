# Correlation Factor and Variance Inflation Factor in Python

```python
import numpy as np
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

Dataset: UCI Bank Marketing Dataset (https://archive.ics.uci.edu/ml/datasets/bank+marketing)

```python
df = pd.read_csv('data/bank-additional-full.csv')
df.head()
```

```python
df.isna().sum().sum()
```

```python
corr_matrix = df.corr().abs()
high_corr_var=np.where(corr_matrix>0.8)
high_corr_var=[(corr_matrix.columns[x],corr_matrix.columns[y]) for x,y in zip(*high_corr_var) if x!=y and x<y]
high_corr_var
```

```python
df.describe()
```

```python
vif_df = df[['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']]
vif_df.head()
```

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(vif_df.values, i) for i in range(vif_df.shape[1])]
vif["features"] = vif_df.columns
vif.sort_values(by=['VIF Factor'])
```

## From correlation factor and variance inflation factor, we can say that it is safe to delete euribor3m to remove severe multi-collinearity from the dataset.
